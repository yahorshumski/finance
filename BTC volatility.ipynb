```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import datetime
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
#import plotly.io as pio


pd.options.plotting.backend = "plotly"
#pd.options.mode.chained_assignment = None  # default='warn'

start = datetime.datetime(2016, 6, 27)
end = datetime.datetime(2021, 6, 27)

data_provider = 'quandl' #  or fred or quandl
ticker = 'BCHAIN/MKPRU'
BTC = web.DataReader(ticker, data_provider, start, end, api_key = 'PpZt3Tx8SqDzs15WbFex')
BTCUSD = BTC[::-1]
```

```python
BTCUSD
```

```python
returns = np.log(BTCUSD/BTCUSD.shift(1))
#returns = (BTCUSD/BTCUSD.shift(1))-1

returns.fillna(0, inplace=True)
volatility = returns.rolling(window=90).std()*np.sqrt(365)
```

```python

y2 = go.layout.YAxis(overlaying='y', side='right', title= '3 months annualized volatility', color='red')
layout = go.Layout(yaxis2=y2)

fig = go.Figure(layout_title_text="BTC price vs Standard deviation", layout_title_x=0.5, layout=go.Layout(yaxis2=y2))

fig.add_scatter(x=BTCUSD.index, y = BTCUSD['Value'], mode="lines", name="BTCUSD")
fig.add_scatter(x=volatility.index, y = volatility['Value'], mode="lines", name="stdev", yaxis='y2')

#line=dict(color="navy")

fig.add_layout_image(
        dict(
            source="https://images.plot.ly/language-icons/api-home/python-logo.png",
            xref="x",
            yref="y",
            x=0,
            y=3,
            sizex=2,
            sizey=2,
            sizing="stretch",
            opacity=0.5,
            layer="below")
)

fig.update_layout(
    autosize=False,
    width=900,
    height=700,
    #template='xgridoff',
)


#fig.update_yaxes(type="log")

fig.show()
```
