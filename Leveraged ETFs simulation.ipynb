```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import plotly.express as px
import plotly.graph_objs as go
import yfinance as yf

pd.options.plotting.backend = "plotly"
pd.options.mode.chained_assignment = None  # default='warn'

#start = datetime.datetime(2020, 9, 1)
#end = datetime.datetime(2021, 6, 10)

#data_provider = 'google' #  or fred or quandl
ticker1 = yf.Ticker('SPY')  # 'MULTPL/SP500_REAL_PRICE_MONTH'
#SP500 = web.DataReader(ticker, data_provider, start, end, api_key = 'PpZt3Tx8SqDzs15WbFex')
SP = ticker1.history(start="2007-01-01", end="2018-01-01")
UPRO = yf.Ticker('UPRO').history(start="2007-01-01", end="2018-01-01")
SSO = yf.Ticker('SSO').history(start="2007-01-01", end="2018-01-01")
 

UPRO_ratio = 0.0093
trading_days = 252
```

```python
SP
```

```python
SP['Daily_return'] = pd.Series(0)
SP['Daily_return'] = (SP['Close'] / SP['Close'].shift(1)) - 1
SP.dropna(inplace = True)
```

```python
UPRO['Daily_return'] = pd.Series(0)
UPRO['Daily_return'] = (UPRO['Close'] / UPRO['Close'].shift(1)) - 1
UPRO.dropna(inplace = True)
```

```python
UPRO
```

```python
SSO['Daily_return'] = pd.Series(0)
SSO['Daily_return'] = (SSO['Close'] / SSO['Close'].shift(1)) - 1
SSO.dropna(inplace = True)
```

```python
SSO
```

```python
x3 = SP.copy()
```

```python
x3
```

```python
for i in range(1, len(x3)):
    x3['Close'].iloc[i] = x3['Close'].iloc[i-1] * ((x3['Daily_return'].iloc[i] - UPRO_ratio / trading_days) * 3) + x3['Close'].iloc[i-1]
x3['Daily_return'] = ( x3['Daily_return'] - UPRO_ratio / trading_days ) * 3
```

```python
x3
```

```python
fig = go.Figure(layout_title_text="Close price", layout_title_x=0.5)

fig.add_scatter(x=SP.index, y=SP['Close'], mode="lines", name="SP 500")
fig.add_scatter(x=UPRO.index, y=UPRO['Close'], mode="lines", name="UPRO")
fig.add_scatter(x=x3.index, y=x3['Close'], mode="lines", name="x3")
fig.add_scatter(x=SSO.index, y=SSO['Close'], mode="lines", name="SSO")


fig.update_yaxes(type="log")

fig.show()


#fig = px.line(SP, x=SP.index, y=SP['Close'], title='SPY')

#fig.add_scatter(x=SP.index, y=SP['Close'], mode="lines", name="SP 500")
#fig.add_scatter(x=UPRO.index, y=UPRO['Close'], mode="lines", name="UPRO")
#fig.add_scatter(x=x3.index, y=x3['Close'], mode="lines", name="x3")
#fig.show()

#fig = SP500[['Close', 'x2', 'UPRO']].plot()
#fig.update_layout(yaxis= { 'tickformat': ',.0%'}, hovermode='closest', legend_orientation="h")
#fig = SP500.line(df, x="Date", y=[Close, x2], color='Leverage')
#fig.show()
```

```python
SP['Cumulative_return'] = (1 + SP['Daily_return']).cumprod() - 1
UPRO['Cumulative_return'] = (1 + UPRO['Daily_return']).cumprod() - 1
SSO['Cumulative_return'] = (1 + SSO['Daily_return']).cumprod() - 1
x3['Cumulative_return'] = (1 + x3['Daily_return']).cumprod() - 1
```
